```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
employeesdf=pd.read_csv('csv/employees.csv')
employeesdf.head()
```

```python
dfFirstLastName=employeesdf[['first_name','last_name']]
dfFirstLastName
```

```python
dfFirstLastName["email"]=dfFirstLastName["first_name"]+dfFirstLastName["last_name"]+"@pavasjsu.onmicrosoft.com"
dfFirstLastName
```

```python
dfFirstLastName["Inital_Password"]="myPassword1234"
dfFirstLastName
```

```python
displayname=dfFirstLastName["first_name"]+" "+dfFirstLastName["last_name"]
dfdisplayname=pd.DataFrame(data=displayname)
dfdisplayname.columns=["Name [displayName] Required"]
dfdisplayname["User name [userPrincipalName] Required"]=dfFirstLastName["email"]
dfdisplayname["Initial password [passwordProfile] Required"]=dfFirstLastName["Inital_Password"]
dfdisplayname["Block sign in (Yes/No) [accountEnabled] Required"]="Yes"
dfdisplayname
```
